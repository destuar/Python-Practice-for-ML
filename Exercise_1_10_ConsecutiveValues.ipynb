```python
import pandas as pd
```

```python
# Letâ€™s create the sample DataFrame
df = pd.DataFrame({
  'item':['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M'], 
  'value':[12, 3, 5, 0, 0, 0, 23, 4, 2, 0, 0, 12, 0]
})
df
```

```python
df[df['value'] == 0].groupby((df['value'] != 0).cumsum())
```

```python
for k, v in df[df['value'] == 0].groupby((df['value'] != 0).cumsum()):
    print(f'[group {k}]')
    print(v)
    print('\n')
```
